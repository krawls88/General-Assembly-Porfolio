###### Usings the pandas dataframe I am creating a pivot table based on Passenger/Crew, and Department so I can better understand the rate of survival betwee the different subsets of passengers.  

```python
Lusitania[['Passenger/Crew', 'Department/Class', 'Fate']].pivot_table

L1 = Lusitania.pivot_table('Title', index=['Passenger/Crew', 'Department/Class'], columns=['Fate'], fill_value=0, aggfunc=len, margins=True)
L1.columns


#Calculating the Survival rate as Saved over total number of passengers or represented here as 
#total of (Lost, Saved, Not on Board, Saved (but died from trauma))
L1['Survival Rate'] = L1['Saved'] / (L1['Lost'] + L1['Saved'] + L1['Not on board'])
L1
```

###### Above we can see the Band had the highest survival rate followed by the member fo the Deck.  Overall, the members of the Crew had a higher survial rate than the passengers .

###### In my pivot table above I have a few weird titles Mr. (later Sir), and Master (?).  I am going to remove them, creating a new data set where Mr. (Later Sir) and Master (?) will be joing with Mr. and Master respectively.  

```python
new_Title = []

for i in Lusitania['Title']:
    x = i.split(' ')
    w = x[0].split(',')    
    y = w[0].strip()
    new_Title.append(y)
```

```python
Lusitania['Title'] = new_Title
```

```python
L2 = Lusitania.pivot_table('Family name', index=['Title'], columns=['Passenger/Crew', 'Fate'], fill_value=0, aggfunc=len, margins=True)
```
